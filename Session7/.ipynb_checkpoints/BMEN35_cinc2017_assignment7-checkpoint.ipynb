```python
# You know this now.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

from tensorflow.keras.layers import Conv1D
from tensorflow.keras.layers import MaxPooling1D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten

from tensorflow.keras.optimizers import SGD

from tensorflow.keras.models import Sequential
from tensorflow.keras.models import Model
from tensorflow.keras.models import load_model

from tensorflow.keras.utils import to_categorical
from tensorflow.keras.utils import plot_model

# Maybe add some optimizers and callbacks?
```

```python
feats_train = pd.read_csv('cinc2017_features_training.csv', header=None)
feats_val = pd.read_csv('cinc2017_features_validation.csv', header=None)

raw_train = pd.read_csv('cinc2017_raw_train.csv', header=None)
raw_val = pd.read_csv('cinc2017_raw_validation.csv', header=None)

labels_train = pd.read_csv('cinc2017_labels_training.csv', header=None)
labels_val = pd.read_csv('cinc2017_labels_validation.csv', header=None)
```

```python
X_feats_train = feats_train.to_numpy()
X_feats_val = feats_val.to_numpy()

X_raw_train = raw_train.to_numpy()
X_raw_val = raw_val.to_numpy()

y_labels_train = to_categorical(labels_train)
y_labels_val = to_categorical(labels_val)

labels = ['Normal', 'AF', 'Other', 'Noisy']
```

## Andreotti et al.

```python
andreotti = load_model('andreotti_et_al.h5')

val_results = andreotti.evaluate(X_raw_val, y_labels_val, verbose=0)
print("Validation accuracy for Andreotti is {:.4f} ".format(val_results[1]))

# andreotti.summary()
plot_model(andreotti, show_shapes=True, show_layer_names=True, show_layer_activations=True)
```

```python
predictions = andreotti.predict(X_raw_val)
```

```python
y_pred = np.argmax(predictions, axis=1)
print(classification_report(labels_val, y_pred, target_names=labels))
```

Here we want to look at the "macro avg" for the "f1-score" which is 0.86.

## "Deep learning" model

```python
model = Sequential()
# We will add some layers to this model
model.add(Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(18000, 1)))
model.add(Flatten())
model.add(Dense(4, activation='softmax'))
# compile model
opt = SGD(learning_rate=0.01, momentum=0.9)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
plot_model(model, show_shapes=True, show_layer_names=True, show_layer_activations=True) # This also generates a file of the plot
```

```python
model.fit(X_raw_train, y_labels_train, epochs=10, batch_size=32, verbose=1)
```

```python
preds = model.predict(X_raw_val)

y_pred = np.argmax(preds, axis=1)
print(classification_report(labels_val, y_pred, target_names=labels))
```

macro avg = 0.89

## Feature based models

### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold, cross_val_score


rf = RandomForestClassifier(n_estimators = 1500, max_depth = 5)

scaler = StandardScaler()

X_feats_train = scaler.fit_transform(X_feats_train)
X_feats_val = scaler.transform(X_feats_val)


k_folds = KFold(n_splits=5)
score_rf = cross_val_score(rf, X_feats_train, y_labels_train, cv=k_folds)
print(score_rf)
print("Mean of " + str(len(score_rf)) + " folds CV: " + str(np.mean(score_rf)))
```

```python
rf.fit(X_feats_train, y_labels_train)
y_pred = rf.predict(X_feats_val)

print(classification_report(y_labels_val, y_pred, target_names = labels))
```

macro avg = 0.65
